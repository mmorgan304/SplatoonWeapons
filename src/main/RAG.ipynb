```python
import csv
import json

# Make sure this matches the name of your CSV file!
csv_file_path = '/shadow_realm/sql dumps/weapons_w_subs_specials.csv'
json_file_path = 'weapons.json'

knowledge_base = {}

try:
    with open(csv_file_path, mode='r', encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)
        
        for row in reader:
            weapon_name = row['weapon_name']
            
            # Build the blank template for each weapon using the CSV data
            knowledge_base[weapon_name] = {
                "sub": row['subweapon_name'],
                "special": row['special_weapon_name'],
                "strengths": "",
                "weaknesses": "",
                "playstyle": ""
            }

    # Write the dictionary to a beautifully formatted JSON file
    with open(json_file_path, mode='w', encoding='utf-8') as json_file:
        json.dump(knowledge_base, json_file, indent=4)

    print(f"✨ Success! Generated {json_file_path} with {len(knowledge_base)} weapons.")

except FileNotFoundError:
    print(f"❌ Error: Could not find '{csv_file_path}'. Make sure it's in the same folder as this script!")
except KeyError as e:
    print(f"❌ Error: Could not find column {e} in the CSV. Check your CSV headers!")
```
