```python
import csv
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
import os
import json
import gc # ✨ ADDED: Garbage Collection module for RAM management

# ==========================================
# 1. PARSE CSV & BUILD WEAPON KITS DICTIONARY
# ==========================================
param_df = pd.read_csv("sql dumps/sp_weapons_complete.csv", encoding="utf-8")
param_df['firing_modes'] = param_df['firing_modes'].astype(str)

param_df['parsed_firing_modes'] = param_df['firing_modes'].apply(lambda x: json.loads(x) if pd.notna(x) and x.startswith('[') else None)
firing_modes_flat = param_df.explode('parsed_firing_modes')[['weapon_name', 'parsed_firing_modes']].dropna()
firing_modes_expanded = pd.json_normalize(firing_modes_flat['parsed_firing_modes']).set_index(firing_modes_flat.index).join(firing_modes_flat['weapon_name'])
firing_modes_expanded['capped_maxDamage'] = np.minimum(firing_modes_expanded['maxDamage'], 100)

firing_modes_agg = firing_modes_expanded.groupby('weapon_name').agg(
    max_damage_agg=('capped_maxDamage', 'max'),
    min_damage_agg=('minDamage', 'min'),
    ink_efficiency_agg=('inkEfficiency', 'mean'),
    effective_range_agg=('effectiveRange', 'mean'),
    has_oneshot_agg=('hasOneshot', 'max')
).reset_index()

param_df = pd.merge(param_df, firing_modes_agg, on='weapon_name', how='left')
param_df['has_oneshot_agg'] = param_df['has_oneshot_agg'].fillna(False)

numeric_agg_cols_fill = ['max_damage_agg', 'min_damage_agg', 'ink_efficiency_agg', 'effective_range_agg']
for col in numeric_agg_cols_fill:
    param_df[col] = param_df[col].fillna(param_df[col].mean())

cols_to_numeric = [
    "subweapon_id", "special_weapon_id", "weight_id", "special_points",
    "range_actual", "avg_paint_per_match"
]
numeric_agg_cols = ['max_damage_agg', 'min_damage_agg', 'ink_efficiency_agg', 'effective_range_agg']
cols_to_numeric.extend(numeric_agg_cols)

for col in cols_to_numeric:
    param_df[col] = pd.to_numeric(param_df[col], errors="coerce")

param_df["weapon_name"] = param_df["weapon_name"].str.strip()
param_df["secret_name"] = param_df["secret_name"].str.strip()

all_subs = sorted(param_df["subweapon_id"].dropna().unique().astype(int))
all_specials = sorted(param_df["special_weapon_id"].dropna().unique().astype(int))

defaults = {
    "avg_paint": param_df["avg_paint_per_match"].mean(),
    "range": param_df["range_actual"].mean(),
    "special_pts": param_df["special_points"].mean(),
    "weight": param_df["weight_id"].mean(),
    "sub_id": -1,
    "special_id": -1,
    "max_damage": param_df["max_damage_agg"].mean(),
    "min_damage": param_df["min_damage_agg"].mean(),
    "ink_efficiency": param_df["ink_efficiency_agg"].mean(),
    "effective_range": param_df["effective_range_agg"].mean(),
    "has_oneshot": False,
}

kit_dict = {}
for _, row in param_df.iterrows():
    stats = {
        "avg_paint": row["avg_paint_per_match"],
        "mm_range": row["matchmaking_range"],
        "range": row["range_actual"],
        "special_pts": row["special_points"],
        "weight": row["weight_id"],
        "sub_id": int(row["subweapon_id"]),
        "special_id": int(row["special_weapon_id"]),
        "max_damage": row["max_damage_agg"],
        "min_damage": row["min_damage_agg"],
        "ink_efficiency": row["ink_efficiency_agg"],
        "effective_range": row["effective_range_agg"],
        "has_oneshot": row["has_oneshot_agg"],
    }
    kit_dict[row["secret_name"]] = stats
    kit_dict[row["weapon_name"]] = stats

# ==========================================
# 2. LOAD AND SPLIT BASE SPLATOON MATCH DATA
# ==========================================
data_frames = []
target_date = "2026-01-29.csv"

for filename in os.listdir('pandas test folder/battle_results'):
    # Check that it matches the YYYY-MM-DD.csv format
    if filename.endswith('.csv') and len(filename) == 14 and filename[4] == '-' and filename[7] == '-':
        
        # Compare the filename directly to our target date
        if filename >= target_date:
            try:
                df = pd.read_csv(os.path.join('pandas test folder/battle_results', filename), on_bad_lines='skip')
                data_frames.append(df)
            except Exception as e:
                print(f"Error reading {filename}: {e}")

splatoon_data = pd.concat(data_frames, ignore_index=True)
del data_frames # ✨ Clean intermediate lists
gc.collect()

print(f"Number of rows in splatoon_data: {len(splatoon_data)}")
weapon_cols = ["A1-weapon", "A2-weapon", "A3-weapon", "A4-weapon", "B1-weapon", "B2-weapon", "B3-weapon", "B4-weapon"]
splatoon_data = splatoon_data.dropna(subset=["alpha-count", "bravo-count"] + weapon_cols)

# ✨ DOWNCAST TO INT8 (1 byte instead of 8)
y = (splatoon_data["alpha-count"] > splatoon_data["bravo-count"]).astype(np.int8)
X_context = pd.get_dummies(splatoon_data[['mode', 'stage']], dtype=np.int8)

all_weapons = [w for w in pd.unique(splatoon_data[weapon_cols].values.ravel()) if pd.notna(w)]

# ✨ DOWNCAST DATAFRAMES TO INT8 UPON CREATION
alpha_counts = pd.DataFrame(0, index=splatoon_data.index, columns=[f"Alpha_has_{w}" for w in all_weapons], dtype=np.int8)
bravo_counts = pd.DataFrame(0, index=splatoon_data.index, columns=[f"Bravo_has_{w}" for w in all_weapons], dtype=np.int8)

alpha_slots = ["A1-weapon", "A2-weapon", "A3-weapon", "A4-weapon"]
bravo_slots = ["B1-weapon", "B2-weapon", "B3-weapon", "B4-weapon"]

for weapon in all_weapons:
    for slot in alpha_slots:
        alpha_counts[f"Alpha_has_{weapon}"] += (splatoon_data[slot] == weapon).astype(np.int8)
    for slot in bravo_slots:
        bravo_counts[f"Bravo_has_{weapon}"] += (splatoon_data[slot] == weapon).astype(np.int8)

# ==========================================
# 3. COMPUTE DYNAMIC SUB and SPECIAL
# ==========================================
print("Deconstructing weapon rosters into sub and special combinations...")

def process_match_kits(df, slots, all_subs, all_specials):
    # ✨ DOWNCAST MATRICES TO INT8
    sub_mat = np.zeros((len(df), len(all_subs)), dtype=np.int8)
    spec_mat = np.zeros((len(df), len(all_specials)), dtype=np.int8)

    sub_lookup = {s: i for i, s in enumerate(all_subs)}
    spec_lookup = {sp: i for i, sp in enumerate(all_specials)}

    paint_vals, max_ranges, spec_points, weights = [], [], [], []
    max_damages, min_damages, ink_efficiencies, effective_ranges, has_oneshot_flags = [], [], [], [], []

    # ✨ MASSIVE SPEED/RAM UPGRADE: Loop over raw numpy array instead of pd.iterrows()
    slot_data = df[slots].values

    for row_idx, row_weapons in enumerate(slot_data):
        p_tot, r_max, s_tot, w_tot, mmr_max = 0, 0, 0, 0, 0
        md_max, md_min, ie_tot, er_max, ho_flag = -np.inf, np.inf, 0, -np.inf, False

        num_weapons_in_kit = 0
        for weapon in row_weapons:
            stats = kit_dict.get(weapon, defaults)
            p_tot += stats["avg_paint"]
            s_tot += stats["special_pts"]
            w_tot += stats["weight"]
            if stats["range"] > r_max: r_max = stats["range"]
            if stats["mm_range"] > mmr_max: mmr_max = stats["mm_range"]

            if stats["max_damage"] > md_max: md_max = stats["max_damage"]
            if stats["min_damage"] < md_min: md_min = stats["min_damage"]
            ie_tot += stats["ink_efficiency"]
            if stats["effective_range"] > er_max: er_max = stats["effective_range"]
            if stats["has_oneshot"]: ho_flag = True
            num_weapons_in_kit += 1

            if stats["sub_id"] in sub_lookup:
                sub_mat[row_idx, sub_lookup[stats["sub_id"]]] += 1
            if stats["special_id"] in spec_lookup:
                spec_mat[row_idx, spec_lookup[stats["special_id"]]] += 1

        paint_vals.append(p_tot)
        max_ranges.append(r_max)
        spec_points.append(s_tot)
        weights.append(w_tot / float(num_weapons_in_kit) if num_weapons_in_kit > 0 else 0)

        max_damages.append(md_max if md_max != -np.inf else 0)
        min_damages.append(md_min if md_min != np.inf else 0)
        ink_efficiencies.append(ie_tot / float(num_weapons_in_kit) if num_weapons_in_kit > 0 else 0)
        effective_ranges.append(er_max if er_max != -np.inf else 0)
        has_oneshot_flags.append(ho_flag)

    return sub_mat, spec_mat, paint_vals, max_ranges, spec_points, weights, \
           max_damages, min_damages, ink_efficiencies, effective_ranges, has_oneshot_flags


# Extract Alpha & Bravo Metrics
a_sub, a_spec, a_paint, a_range, a_pts, a_weight, a_mm_range, \
a_max_damage, a_min_damage, a_ink_efficiency, a_effective_range, a_has_oneshot = process_match_kits(
    splatoon_data, alpha_slots, all_subs, all_specials
)
b_sub, b_spec, b_paint, b_range, b_pts, b_weight, b_mm_range, \
b_max_damage, b_min_damage, b_ink_efficiency, b_effective_range, b_has_oneshot = process_match_kits(
    splatoon_data, bravo_slots, all_subs, all_specials
)

# ✨ FREE RAM EARLY: splatoon_data is no longer needed after processing kits
del splatoon_data
gc.collect()

# Convert arrays directly to strategic team differentials (Alpha - Bravo)
# ✨ CAST CONTINUOUS DATA TO FLOAT32
df_diffs = pd.DataFrame()
df_diffs["Diff_total_paint"] = (np.array(a_paint) - np.array(b_paint)).astype(np.float32)
df_diffs["Diff_max_range"] = (np.array(a_range) - np.array(b_range)).astype(np.float32)
df_diffs["Diff_total_spec_pts"] = (np.array(a_pts) - np.array(b_pts)).astype(np.float32)
df_diffs["Diff_mean_weight"] = (np.array(a_weight) - np.array(b_weight)).astype(np.float32)
df_diffs["Diff_max_damage"] = (np.array(a_max_damage) - np.array(b_max_damage)).astype(np.float32)
df_diffs["Diff_min_damage"] = (np.array(a_min_damage) - np.array(b_min_damage)).astype(np.float32)
df_diffs["Diff_ink_efficiency"] = (np.array(a_ink_efficiency) - np.array(b_ink_efficiency)).astype(np.float32)
df_diffs["Diff_effective_range"] = (np.array(a_effective_range) - np.array(b_effective_range)).astype(np.float32)

# ✨ CAST DISCRETE DATA TO INT8
df_diffs["Diff_has_oneshot"] = (np.array(a_has_oneshot).astype(np.int8) - np.array(b_has_oneshot).astype(np.int8))

for i, s in enumerate(all_subs):
    df_diffs[f"Diff_count_sub_{s}"] = (a_sub[:, i] - b_sub[:, i]).astype(np.int8)
for i, sp in enumerate(all_specials):
    df_diffs[f"Diff_count_spec_{sp}"] = (a_spec[:, i] - b_spec[:, i]).astype(np.int8)

# Merge everything together (Reset index to align perfectly since splatoon_data was dropped)
X_context.reset_index(drop=True, inplace=True)
alpha_counts.reset_index(drop=True, inplace=True)
bravo_counts.reset_index(drop=True, inplace=True)
X_encoded = pd.concat([X_context, alpha_counts, bravo_counts, df_diffs], axis=1)

del X_context, alpha_counts, bravo_counts, df_diffs
gc.collect()

# ==========================================
# 3.5 MIRROR THE DATASET FOR SYMMETRY
# ==========================================
print("Mirroring dataset to enforce faction symmetry...")

alpha_cols = [col for col in X_encoded.columns if col.startswith('Alpha_has_')]
bravo_cols = [col for col in X_encoded.columns if col.startswith('Bravo_has_')]
diff_cols = [col for col in X_encoded.columns if col.startswith('Diff_')]

X_mirrored = X_encoded.copy()
X_mirrored[alpha_cols] = X_encoded[bravo_cols].values
X_mirrored[bravo_cols] = X_encoded[alpha_cols].values
X_mirrored[diff_cols] = -X_encoded[diff_cols]

y_mirrored = 1 - y

X_doubled = pd.concat([X_encoded, X_mirrored], ignore_index=True)
y_doubled = pd.concat([y, y_mirrored], ignore_index=True)

print(f"Original dataset size: {len(X_encoded)}")
print(f"Doubled dataset size: {len(X_doubled)}")

# Store columns before deleting the large DataFrame
trained_features_columns = X_doubled.columns

# ✨ EXTREME GARBAGE COLLECTION: Delete the 4 original shards before passing into the ML step
del X_encoded, X_mirrored, y, y_mirrored
gc.collect()

# ==========================================
# 4. TRAIN AND ASSESS MODEL
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(
    X_doubled, y_doubled, test_size=0.2, random_state=42
)

# Once the split is done, delete the massive doubled arrays from memory
del X_doubled, y_doubled
gc.collect()

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1) # n_jobs=-1 speeds up training
rf_classifier.fit(X_train, y_train)

print(f"\n✨ Kit-Aware Match Prediction Accuracy: {rf_classifier.score(X_test, y_test):.2%}")
```

```python
import joblib
import pandas as pd
import numpy as np

# Load your pre-trained model
print("Loading model...")
rf_classifier = joblib.load('src/main/python/rf_splatoon_model.joblib')
print("Model loaded and ready!")

trained_features_columns = rf_classifier.feature_names_in_


def find_optimal_team_comp_v3(
    model,
    trained_features,
    weapon_pools_per_slot,
    rich_kit_dict,
    fallback_defaults,
    all_subs,
    all_specials,
    target_mode,
    target_stage,
    bravo_team=None,
    max_iter=9
):
    """
    Optimizes a 4-weapon team using Iterative Coordinate Ascent.
    Ensures the ML model always evaluates a complete 4v4 layout, keeping
    all differential metrics perfectly in-distribution.
    """
    # 1. FIXED INITIALIZATION: Pull the first valid weapon from EACH slot's specific pool
    optimized_alpha_team = []
    for pool in weapon_pools_per_slot:
        optimized_alpha_team.append(pool[0] if pool else "liter4k_scope")

    # Build base match scenario dataframe
    match_scenario_base = pd.DataFrame(0, index=[0], columns=trained_features)
    mode_col, stage_col = f"mode_{target_mode}", f"stage_{target_stage}"

    if mode_col not in match_scenario_base.columns or stage_col not in match_scenario_base.columns:
        print(f"❌ Error: Could not find contexts in trained model columns.")
        return None, None

    match_scenario_base[mode_col] = 1
    match_scenario_base[stage_col] = 1

    # Pre-calculate baseline metrics for the opponent team (Bravo)
    b_paint, b_range, b_pts, b_weight = 0, 0, 0, 0
    b_max_damage, b_min_damage, b_ink_efficiency, b_effective_range, b_has_oneshot = -np.inf, np.inf, 0, -np.inf, False
    b_sub_counts = {s: 0 for s in all_subs}
    b_spec_counts = {sp: 0 for sp in all_specials}

    if bravo_team:
        for weapon in bravo_team:
            bravo_col = f"Bravo_has_{weapon}"
            if bravo_col in match_scenario_base.columns:
                match_scenario_base[bravo_col] += 1

            stats = rich_kit_dict.get(weapon, fallback_defaults)
            b_paint += stats["avg_paint"]
            b_pts += stats["special_pts"]
            b_weight += stats["weight"]
            if stats["range"] > b_range:
                b_range = stats["range"]
            if stats["max_damage"] > b_max_damage:
                b_max_damage = stats["max_damage"]
            if stats["min_damage"] < b_min_damage:
                b_min_damage = stats["min_damage"]
            b_ink_efficiency += stats["ink_efficiency"]
            if stats["effective_range"] > b_effective_range:
                b_effective_range = stats["effective_range"]
            if stats["has_oneshot"]:
                b_has_oneshot = True

            if stats["sub_id"] in b_sub_counts:
                b_sub_counts[stats["sub_id"]] += 1
            if stats["special_id"] in b_spec_counts:
                b_spec_counts[stats["special_id"]] += 1

    print(f"🔮 Optimizing composition for {target_mode.upper()} on {target_stage} via Coordinate Ascent...")
    print("-" * 85)

    overall_best_probability = -1

    # 2. Coordinate Ascent Optimization Loop
    for iteration in range(max_iter):
        has_changed = False

        for slot in range(4):
            # FIXED: Reset the probability tracker for THIS specific slot's evaluation
            best_prob_for_slot = -1
            best_weapon_for_slot = optimized_alpha_team[slot]

            # Use the specific weapon pool for the current slot
            for weapon in weapon_pools_per_slot[slot]:
                # Swap out just the current slot to evaluate a candidate weapon
                proposed_team = list(optimized_alpha_team)
                proposed_team[slot] = weapon

                test_scenario = match_scenario_base.copy()

                # Populate raw weapon markers for Alpha
                for w in proposed_team:
                    alpha_col = f"Alpha_has_{w}"
                    if alpha_col in test_scenario.columns:
                        test_scenario[alpha_col] += 1

                # Tally metrics for this complete 4-weapon candidate composition
                a_paint, a_range, a_pts, a_weight = 0, 0, 0, 0
                a_max_damage, a_min_damage, a_ink_efficiency, a_effective_range, a_has_oneshot = -np.inf, np.inf, 0, -np.inf, False

                a_sub_counts = {s: 0 for s in all_subs}
                a_spec_counts = {sp: 0 for sp in all_specials}

                for w in proposed_team:
                    stats = rich_kit_dict.get(w, fallback_defaults)
                    a_paint += stats["avg_paint"]
                    a_pts += stats["special_pts"]
                    a_weight += stats["weight"]
                    if stats["range"] > a_range:
                        a_range = stats["range"]
                    if stats["max_damage"] > a_max_damage:
                        a_max_damage = stats["max_damage"]
                    # if stats["min_damage"] < a_min_damage:
                    #     a_min_damage = stats["min_damage"]
                    a_ink_efficiency += stats["ink_efficiency"]
                    if stats["effective_range"] > a_effective_range:
                        a_effective_range = stats["effective_range"]
                    if stats["has_oneshot"]:
                        a_has_oneshot = True

                    if stats["sub_id"] in a_sub_counts:
                        a_sub_counts[stats["sub_id"]] += 1
                    if stats["special_id"] in a_spec_counts:
                        a_spec_counts[stats["special_id"]] += 1

                # Apply Differentials
                test_scenario["Diff_total_paint"] = a_paint - b_paint
                test_scenario["Diff_max_range"] = a_range - b_range
                test_scenario["Diff_total_spec_pts"] = a_pts - b_pts
                test_scenario["Diff_mean_weight"] = (a_weight / 4.0) - (b_weight / 4.0 if bravo_team else 0)
                test_scenario["Diff_max_damage"] = a_max_damage - b_max_damage
                # test_scenario["Diff_min_damage"] = a_min_damage - b_min_damage
                test_scenario["Diff_ink_efficiency"] = a_ink_efficiency - b_ink_efficiency
                test_scenario["Diff_effective_range"] = a_effective_range - b_effective_range
                test_scenario["Diff_has_oneshot"] = a_has_oneshot - b_has_oneshot

                for s in all_subs:
                    col = f"Diff_count_sub_{s}"
                    if col in test_scenario.columns:
                        test_scenario[col] = a_sub_counts[s] - b_sub_counts[s]
                for sp in all_specials:
                    col = f"Diff_count_spec_{sp}"
                    if col in test_scenario.columns:
                        test_scenario[col] = a_spec_counts[sp] - b_spec_counts[sp]

                # Query model probability
                win_probability = model.predict_proba(test_scenario)[0][1]

                # Compare against THIS slot's best probability
                if win_probability > best_prob_for_slot:
                    best_prob_for_slot = win_probability
                    best_weapon_for_slot = weapon

            # If a better weapon was found for this slot, update the team formation
            if optimized_alpha_team[slot] != best_weapon_for_slot:
                optimized_alpha_team[slot] = best_weapon_for_slot
                has_changed = True

            # Keep track of the final overall team probability
            overall_best_probability = best_prob_for_slot

        print(f"🔄 Pass {iteration + 1} Complete -> Current Roster: {optimized_alpha_team} | Projected Win Rate: {overall_best_probability:.2%}")

        # If an entire sweep happens without changing a single weapon, the composition is fully maximized
        if not has_changed:
            print("✨ Team composition stabilized. Halting optimization execution.")
            break

    print("-" * 85)
    print(f"✅ Final Kit-Aware Optimized Composition: {optimized_alpha_team}")
    return optimized_alpha_team, overall_best_probability

# ==========================================
# EXECUTE COUNTER-PICK SELECTION SIMULATION
# ==========================================
target_mode = "area"
target_stage = "taraport"
enemy_comp = ["liter4k", "barrelspinner", "hotblaster", "52gal"]

weapon_pool_bad = [
    "wideroller", "moprin", "moprin_kaku", "squiclean_b", "sputtery", "sputtery_hue", "carbon_deco", "carbon_angl", "pablo", "hokusai", "hokusai_hue", "hokusai_sui"
]

weapon_pool_top_120 = [
    "52gal", "sshooter", "liter4k", "hotblaster_custom", "sshooter_collabo",
    "dynamo", "wakaba", "nova_neo", "barrelspinner", "furo",
    "bold", "nzap85", "sputtery_hue", "carbon_deco", "splatroller",
    "maneuver_collabo", "prime_collabo", "promodeler_rg", "promodeler_mg", "momiji",
    "sharp", "hotblaster", "liter4k_scope", "hydra", "pablo",
    "hissen_hue", "order_shooter_replica", "prime_frzn", "hokusai_hue", "hydra_atsu",
    "nzap89", "tristringer_tou", "furuido", "tristringer", "fincent",
    "sharp_neo", "lact450_milk", "quadhopper_black", "nautilus47", "explosher_custom",
    "bucketslosher", "order_spinner_replica", "jetsweeper_cobr", "order_roller_replica", "variableroller",
    "examiner", "prime", "moprin", "dualsweeper", "liter4k_custom",
    "fincent_hue", "squiclean_a", "rapid_elite_deco", "heroshooter_replica", "jimuwiper",
    "maneuver", "sputtery_owl", "dynamo_mei", "pablo_hue", "sblast91",
    "hissen_ash", "bold_neo", "nautilus79", "furuido_custom", "octoshooter_replica",
    "longblaster", "clashblaster", "dentalwiper_mint", "tristringer_collabo", "spygadget",
    "rpen_5h", "sputtery", "dynamo_tesla", "explosher", "hokusai",
    "96gal", "52gal_deco", "splatscope", "drivewiper", "wideroller_waku",
    "sharp_geck", "sshooter_kou", "splatspinner", "hokusai_sui", "jetsweeper_custom",
    "splatcharger", "kelvin525", "splatscope_collabo", "spaceshooter", "hissen",
    "liter4k_scope_custom", "fincent_brnz", "order_slosher_replica", "clashblaster_neo", "screwslosher",
    "kugelschreiber", "order_maneuver_replica", "rapid_elite_wntr", "nova", "rapid_deco",
    "parashelter_sorella", "spaceshooter_collabo", "lact450_deco", "jimuwiper_fuu", "jetsweeper",
    "lact450", "longblaster_custom", "order_stringer_replica", "splatcharger_collabo", "variableroller_foil",
    "bottlegeyser", "campingshelter_sorella", "96gal_sou", "bamboo14mk1", "promodeler_sai",
    "examiner_hue", "barrelspinner_deco", "maneuver_you", "screwslosher_neo", "brella24mk2"
]

weapon_pool_top_80  = [
    "52gal", "sshooter", "liter4k", "hotblaster_custom", "sshooter_collabo",
    "dynamo", "wakaba", "nova_neo", "barrelspinner", "furo",
    "bold", "nzap85", "sputtery_hue", "carbon_deco", "splatroller",
    "maneuver_collabo", "prime_collabo", "promodeler_rg", "promodeler_mg", "momiji",
    "sharp", "hotblaster", "liter4k_scope", "hydra", "pablo",
    "hissen_hue", "order_shooter_replica", "prime_frzn", "hokusai_hue", "hydra_atsu",
    "nzap89", "tristringer_tou", "furuido", "tristringer", "fincent",
    "sharp_neo", "lact450_milk", "quadhopper_black", "nautilus47", "explosher_custom",
    "bucketslosher", "order_spinner_replica", "jetsweeper_cobr", "order_roller_replica", "variableroller",
    "examiner", "prime", "moprin", "dualsweeper", "liter4k_custom",
    "fincent_hue", "squiclean_a", "rapid_elite_deco", "heroshooter_replica", "jimuwiper",
    "maneuver", "sputtery_owl", "dynamo_mei", "pablo_hue", "sblast91",
    "hissen_ash", "bold_neo", "nautilus79", "furuido_custom", "octoshooter_replica",
    "longblaster", "clashblaster", "dentalwiper_mint", "tristringer_collabo", "spygadget",
    "rpen_5h", "sputtery", "dynamo_tesla", "explosher", "hokusai",
    "96gal", "52gal_deco", "splatscope", "drivewiper", "wideroller_waku"
]

weapon_pool_top_40 = [
    "52gal", "sshooter", "liter4k", "hotblaster_custom", "sshooter_collabo",
    "dynamo", "wakaba", "nova_neo", "barrelspinner", "furo",
    "bold", "nzap85", "sputtery_hue", "carbon_deco", "splatroller",
    "maneuver_collabo", "prime_collabo", "promodeler_rg", "promodeler_mg", "momiji",
    "sharp", "hotblaster", "liter4k_scope", "hydra", "pablo",
    "hissen_hue", "order_shooter_replica", "prime_frzn", "hokusai_hue", "hydra_atsu",
    "nzap89", "tristringer_tou", "furuido", "tristringer", "fincent",
    "sharp_neo", "lact450_milk", "quadhopper_black", "nautilus47", "explosher_custom"
]

# --- Find Maximum Win Rate ---
print("\n--- Calculating Maximum Win Rate ---")
weapon_pools_per_slot_custom_max = [
    weapon_pool_top_40,  # Player 1 (slot 0)
    weapon_pool_top_80,  # Player 2 (slot 1)
    weapon_pool_top_120,  # Player 3 (slot 2)
    weapon_pool_top_120   # Player 4 (slot 3)
]

best_team_max, max_win_rate = find_optimal_team_comp_v3(
    model=rf_classifier,
    trained_features=trained_features_columns, # ✨ FIXED: Swapped out X_doubled.columns
    weapon_pools_per_slot=weapon_pools_per_slot_custom_max,
    rich_kit_dict=kit_dict,
    fallback_defaults=defaults,
    all_subs=all_subs,
    all_specials=all_specials,
    target_mode=target_mode,
    target_stage=target_stage,
    bravo_team=enemy_comp,
)

# --- Find Minimum Win Rate (Terrible Team Comp with Duplicates Allowed) ---
print("\n--- Calculating Minimum Win Rate (Terrible Team Comp) ---")
weapon_pools_per_slot_custom_min = [
    ["soytuber"],
    ["soytuber"],
    ["soytuber"],
    ["soytuber"]
]

bad_team_min, min_win_rate = find_optimal_team_comp_v3(
    model=rf_classifier,
    trained_features=trained_features_columns, # ✨ FIXED: Swapped out X_doubled.columns
    weapon_pools_per_slot=weapon_pools_per_slot_custom_min,
    rich_kit_dict=kit_dict,
    fallback_defaults=defaults,
    all_subs=all_subs,
    all_specials=all_specials,
    target_mode=target_mode,
    target_stage=target_stage,
    bravo_team=enemy_comp,
)

print("\n--- Calculating Equal Win Rate (Same Team Comp) ---")
weapon_pools_per_slot_equal = [
    ["liter4k"],        # ✨ FIXED: Lowercase 'k' to match extraction keys
    ["barrelspinner"],
    ["hotblaster"],
    ["52gal"]
]

equal_team_min, equal_win_rate = find_optimal_team_comp_v3(
    model=rf_classifier,
    trained_features=trained_features_columns, # ✨ FIXED: Swapped out X_doubled.columns
    weapon_pools_per_slot=weapon_pools_per_slot_equal,
    rich_kit_dict=kit_dict,
    fallback_defaults=defaults,
    all_subs=all_subs,
    all_specials=all_specials,
    target_mode=target_mode,
    target_stage=target_stage,
    bravo_team=enemy_comp,
)
```
